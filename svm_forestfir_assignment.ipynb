```python
# convert data :
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["month"] = le.fit_transform(data["month"])
data["day"] = le.fit_transform(data["day"])
data["size_category"] = le.fit_transform(data["size_category"])
data.head()
```

```python
# split data :
array = data.values
x = array[:,0:30]
y = array[:,30]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 7)
x_train.shape, x_test.shape
```

```python
# apply svm without feature selection :
from sklearn.svm import SVC
svm = SVC(kernel = 'rbf', random_state = 0, gamma=0.10, C=1.0)
svm.fit(x_train, y_train)
print("training accuracy :", svm.score(x_train, y_train))            # 100 % training accuracy
print("testing accuracy :", svm.score(x_test, y_test))               # 66.9 % testing accuracy
# using whole attributes we get overfitting problem.
```

```python
# featur selection using Tree Based Feature Selection method ;
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x, y)
print(model.feature_importances_)
l = model.feature_importances_>0.1
l
# using Tree base feature selection we got that only "area" attribute has 100 % impact for classifying size_category.
```

```python
# RFE (feature selection):
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
array = data.values
x = array[:,0:30]
y = array[:,30]
model = LogisticRegression(max_iter=400)
rfe = RFE(model, 3)
fit = rfe.fit(x, y)
print(fit.ranking_)
```
